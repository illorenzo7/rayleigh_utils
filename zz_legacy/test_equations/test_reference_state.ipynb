```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import os, sys

sys.path.append(os.environ['raco'])
sys.path.append(os.environ['rapl'])
sys.path.append(os.environ['rapp'])
from common import *
```

```python
eq = get_eq('.')
```

```python
dlnrho_num = drad(eq.rho, eq.radius)/eq.rho
d2lnrho_num = drad(dlnrho_num, eq.radius)
dlnt = drad(eq.T, eq.radius)/eq.T
dprs = drad(eq.pressure, eq.radius)
```

```python
plt.figure(figsize=(10,6))
plt.plot(eq.radius, eq.d2lnrho, 'k', label='ref')
plt.plot(eq.radius, d2lnrho_num, 'r--', label='num')
plt.legend()
plt.title('d^2 ln rho / dr')
```

```python
plt.figure(figsize=(10,6))
plt.plot(eq.radius, eq.dlnrho, 'k', label='ref')
plt.plot(eq.radius, dlnrho_num, 'r--', label='num')
plt.legend()
plt.title('d ln rho / dr')
```

```python
plt.figure(figsize=(10,6))
plt.plot(eq.radius, -eq.rho*eq.gravity, 'k', label='- rho * g')
plt.plot(eq.radius, dprs, 'r--', label='dP/dr')
plt.legend()
plt.title('hydro balance')
```
