```python
def project_onto_plane(r, theta, phi, alpha=60., beta=25.):
    alpha_rad = alpha*np.pi/180.
    beta_rad = beta*np.pi/180.
    
    points_proj = r*np.array([np.sin(theta)*np.sin(phi - alpha_rad),\
            np.cos(beta_rad)*np.cos(theta) - np.sin(beta_rad)*np.sin(theta)*np.cos(phi-alpha_rad)])

    discriminant = np.cos(beta_rad)*np.sin(theta)*np.cos(alpha_rad - phi) + np.sin(beta)*np.cos(theta)
    i_bad = np.where(discriminant < 0.)
    i_good = np.where(discriminant >= 0.)
```
