```python
%matplotlib inline
import matplotlib 
%matplotlib inline
matplotlib.rc('xtick', labelsize=8) 
matplotlib.rc('ytick', labelsize=8)
#matplotlib.rc('mathtext', fontset='stix')
matplotlib.rcParams['font.size'] = 8
matplotlib.rcParams['axes.titlesize'] = 10
matplotlib.rcParams['figure.dpi'] = 300
matplotlib.rcParams['lines.linewidth'] = 1

import matplotlib.pyplot as plt
import numpy as np
import sys, os
sys.path.append('./rayleigh/plot')
from rayleigh_diagnostics import Shell_Slices, ReferenceState
from matplotlib import ticker
from matplotlib import colors
from mpl_toolkits.basemap import Basemap, addcyclic
from common import get_widest_range_file, strip_dirname
from sslice_util import plot_ortho, units, labels, var_indices, cbar_axes_from_area
from binormalized_cbar import MidpointNormalize

# Get the run directory on which to perform the analysis
dir5 = '/altair/loma3853/rayleigh/n5_sslice_19100050'
radatadir_vr = dir5 + '/vr/'
radatadir_vp = dir5 + '/vp/'
iter0 = 19100050

# Directory to save plots
datadir = dir5 + '/data/'
plotdir = dir5 + '/plots/'

if not os.path.isdir(plotdir):
    os.makedirs(plotdir)
```

```python
# Get grid info
rr,tt,cost,sint,rr_depth,ri,ro,d = np.load(datadir + 'grid_info.npy')
nt, nr = len(tt), len(rr)
nphi = 2*nt
cost_nd = cost.reshape((1, nt))
sint_nd = sint.reshape((1, nt))
phi = np.arange(0, 2*np.pi, 2*np.pi/nphi)
lats = 90 - tt*180/np.pi 
lons = phi*180/np.pi - 180

# Set dimensions/location of patch to plot 
delta_lat = 40
delta_lon = 40
lon0 = 0
lat0 = 0

ith1, ith2 = np.argmin(np.abs(lats - (lat0 - delta_lat/2))),\
    np.argmin(np.abs(lats - (lat0 + delta_lat/2)))

iph1, iph2 = np.argmin(np.abs(lons - (lon0 - delta_lon/2))),\
    np.argmin(np.abs(lons - (lon0 + delta_lon/2)))
    
xtickvals = [lon0 - delta_lon/2, lon0 - delta_lon/4, lon0, lon0 + delta_lon/4, lon0 + delta_lon/2]
ytickvals = [lat0 - delta_lat/2, lon0 - delta_lat/4, lat0, lat0 + delta_lat/4, lat0 + delta_lat/2]

lonvals = lons[iph1:iph2]
latvals = lats[ith1:ith2]
llons, llats = np.meshgrid(lonvals, latvals, indexing='ij')
```

```python
# Read shell slice for panels
print ('Reading ' + radatadir_vr + str(iter0).zfill(8) + ' ...')
a_vr = Shell_Slices(radatadir_vr + str(iter0).zfill(8), '')
vr = a_vr.vals[:, :, :, 0, 0]/100

print ('Reading ' + radatadir_vp + str(iter0).zfill(8) + ' ...')
a_vp = Shell_Slices(radatadir_vp + str(iter0).zfill(8), '')
vp = a_vp.vals[:, :, :, 0, 0]/100
```

```python
# Compute vrvp fluctuation; trim problematic boundary slices
nshells = len(a_vr.radius)
vr_prime = vr[:, :, 1:-1] - vr_av[:, a_vr.inds[1:-1]].reshape((1, nt, nshells - 2))
vp_prime = vp[:, :, 1:-1] - vp_av[:, a_vp.inds[1:-1]].reshape((1, nt, nshells - 2))

# Compute angular momentum flux
ref = ReferenceState('reference', dir5 + '/')
rho = ref.density
amom_flux = rho[a_vr.inds[1:-1]].reshape((1, 1, nshells - 2))*a_vr.radius[1:-1].reshape((1, 1, nshells - 2))*\
            vr_prime*vp_prime*1e4
```

```python
# Make sure all magnitudes look reasonable
print (np.std(vr), np.std(vr_prime), np.std(vr_av))
print (np.std(vp), np.std(vp_prime), np.std(vp_av))
```

```python
# Read in patches at different depths
min_depth = 0.05
max_depth = 0.5

desired_depths = np.linspace(min_depth, max_depth, 8)

alldepths = (ro - a_vr.radius)/d

depth_indices = []
for desired_depth in desired_depths:
    depth_indices.append(np.argmin(np.abs(alldepths - desired_depth)))
    
rrvals = a_vr.radius[depth_indices]
rrnvals = rrvals/ro

patches = amom_flux[iph1:iph2, ith1:ith2, depth_indices]
```
