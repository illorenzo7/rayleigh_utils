```python
%matplotlib inline
import matplotlib 
%matplotlib inline
matplotlib.rc('xtick', labelsize=6) 
matplotlib.rc('ytick', labelsize=6)
#matplotlib.rc('mathtext', fontset='stix')
matplotlib.rcParams['font.size'] = 6
matplotlib.rcParams['axes.titlesize'] = 8
matplotlib.rcParams['figure.dpi'] = 300
matplotlib.rcParams['lines.linewidth'] = 1

import matplotlib.pyplot as plt
import numpy as np
import sys, os
sys.path.append('./rayleigh/plot')
from rayleigh_diagnostics import Shell_Slices
from matplotlib import ticker
from matplotlib import colors
from mpl_toolkits.basemap import Basemap, addcyclic
from common import get_widest_range_file, strip_dirname
from sslice_util import plot_ortho, units, labels, var_indices, cbar_axes_from_area, rms
from binormalized_cbar import MidpointNormalize

# Get the run directory on which to perform the analysis
dir3 = '/nobackup/lmatilsk/rayleigh/n3_movie_4M'
radatadir = dir3 + '/Shell_Slices/'
first_iter = 4200050
step_iter = 1150

# Directory to save plots
datadir = dir3 + '/data/'
plotdir = dir3 + '/plots/'

if not os.path.isdir(plotdir):
    os.makedirs(plotdir)

# Get grid info
rr,tt,cost,sint,rr_depth,ri,ro,d = np.load(datadir + 'grid_info.npy')
nt, nr = len(tt), len(rr)
nphi = 2*nt
cost_nd = cost.reshape((1, nt))
sint_nd = sint.reshape((1, nt))
phi = np.arange(0, 2*np.pi, 2*np.pi/nphi)
lats = 90 - tt*180/np.pi 
lons = phi*180/np.pi - 180
```

```python
# Set dimensions/location of patch to plot 
delta_lat = 40
delta_lon = 40
lon0 = 0
lat0 = 0

ith1, ith2 = np.argmin(np.abs(lats - (lat0 - delta_lat/2))),\
    np.argmin(np.abs(lats - (lat0 + delta_lat/2)))

iph1, iph2 = np.argmin(np.abs(lons - (lon0 - delta_lon/2))),\
    np.argmin(np.abs(lons - (lon0 + delta_lon/2)))
    
xtickvals = [lon0 - delta_lon/2, lon0 - delta_lon/4, lon0, lon0 + delta_lon/4, lon0 + delta_lon/2]
ytickvals = [lat0 - delta_lat/2, lon0 - delta_lat/4, lat0, lat0 + delta_lat/4, lat0 + delta_lat/2]

lonvals = lons[iph1:iph2]
latvals = lats[ith1:ith2]
llons, llats = np.meshgrid(lonvals, latvals, indexing='ij')
```

```python
iterr = first_iter
print ('Reading ' + radatadir + str(iterr).zfill(8) + ' ...')
a1 = Shell_Slices(radatadir + str(iterr).zfill(8), '')

iterr = first_iter + step_iter*2
print ('Reading ' + radatadir + str(iterr).zfill(8) + ' ...')
a2 = Shell_Slices(radatadir + str(iterr).zfill(8), '')

iterr = first_iter + 2*step_iter*2
print ('Reading ' + radatadir + str(iterr).zfill(8) + ' ...')
a3 = Shell_Slices(radatadir + str(iterr).zfill(8), '')

iterr = first_iter + 3*step_iter*2
print ('Reading ' + radatadir + str(iterr).zfill(8) + ' ...')
a4 = Shell_Slices(radatadir + str(iterr).zfill(8), '')
```

```python
# How much does the differential rotation advect the flow every 1150 time steps?
# dunno why the hell this isn't working! Just do it manually
vavg_file = get_widest_range_file(datadir, 'vavg')

# Get the average velocities
vr_av, vt_av, vp_av = np.load(datadir + vavg_file)
vr_av /= 100; vt_av /= 100; vp_av /= 100

# average vphi over +/- 20 degrees
tt_lat = 90 - tt*180/np.pi
it1, it2 = np.argmin(np.abs(tt_lat + 20)), np.argmin(np.abs(tt_lat - 20))
vp_low = np.mean(vp_av[it1:it2+1, :], axis=0)
ir0 = a1.inds[idepth3]
vp0 = vp_low[ir0]
vp0
```

```python
delta_t = 100*1150 # time between patches in seconds
delta_rphi = delta_t*vp0 # advection of patches in meters
r0 = a3.radius[idepth3]/100 # location of 5% depth in meters
delta_phi_deflection = delta_rphi/r0
delta_phi_deflection_deg = delta_phi_deflection*180/np.pi
delta_phi_deflection_deg
```

```python
# Compare this amount to the resolution
dphi_deg = 360/nphi
dphi_deg
```

```python
# Set up figure dimensions parameters from scratch
fig_width_inches = 3+7/16
margin_middle_inches_x = 1/4
margin_middle_inches_y = 1/4
margin_right_inches = 1/8
margin_bottom_inches = 3/8
margin_left_inches = 1/2
cbar_area_inches = 3/8
subplot_width_inches = (fig_width_inches - margin_left_inches - margin_middle_inches_x - margin_right_inches)/2
subplot_height_inches = subplot_width_inches*delta_lat/delta_lon
fig_height_inches = margin_bottom_inches + 2*subplot_height_inches + margin_middle_inches_y + cbar_area_inches
fig_aspect = fig_height_inches/fig_width_inches

margin_middle_x = margin_middle_inches_x/fig_width_inches
margin_middle_y = margin_middle_inches_y/fig_height_inches
margin_bottom = margin_bottom_inches/fig_height_inches
margin_left = margin_left_inches/fig_width_inches
margin_right = margin_right_inches/fig_width_inches
cbar_area = cbar_area_inches/fig_height_inches
subplot_width = subplot_width_inches/fig_width_inches
subplot_height = subplot_height_inches/fig_height_inches
```

```python
# Plot 5% depth
idepth3 = 0 

# set up colorbar axes
cbar_aspect = 1/20

cbar_width = (1 - margin_left - margin_right)/2
cbar_height= cbar_width*cbar_aspect/fig_aspect

cbar_bottom = 1 - cbar_area/4
cbar_centerx = margin_left + subplot_width + margin_middle_x/2
cbar_left = cbar_centerx - cbar_width/2
```

```python
my_min, my_max = -150, 150
fs, ls = 6, 8
fig = plt.figure(figsize=(fig_width_inches, fig_height_inches))
ax1 = fig.add_axes((margin_left, margin_bottom + subplot_height + margin_middle_y, subplot_width, subplot_height))
ax2 = fig.add_axes((margin_left + subplot_width + margin_middle_x, margin_bottom + subplot_height + margin_middle_y, subplot_width, subplot_height))
ax3 = fig.add_axes((margin_left, margin_bottom, subplot_width, subplot_height))
ax4 = fig.add_axes((margin_left + subplot_width + margin_middle_x, margin_bottom, subplot_width, subplot_height))

cbaxes = fig.add_axes([cbar_left, cbar_bottom, cbar_width, cbar_height])

t1 = a1.time[0]
vr_slice = a1.vals[:, :, idepth3, a1.lut[1], 0]/100
#std_plus = rms(vr_slice[np.where(vr_slice>0)])
#std_minus = rms(vr_slice[np.where(vr_slice<0)])
#my_min, my_max = -3*std_minus, 3*std_plus 
lon0 = 0
iph1, iph2 = np.argmin(np.abs(lons - (lon0 - delta_lon/2))),\
    np.argmin(np.abs(lons - (lon0 + delta_lon/2)))
lonvals = lons[iph1:iph2] - lon0 # Always show -20 to 20 deg. lon but SAY we're in the D.R. frame
llons, llats = np.meshgrid(lonvals, latvals, indexing='ij')

patch = vr_slice[iph1:iph2, ith1:ith2]
plt.sca(ax1)
plt.pcolormesh(llons, llats, patch, cmap=plt.cm.RdYlBu_r, vmin=my_min, vmax=my_max, norm=MidpointNormalize(0))
lw = 0.5
plt.plot(lonvals, np.zeros_like(lonvals), 'k', linewidth=lw)
plt.plot(np.zeros_like(latvals), latvals, 'k', linewidth=lw)
# Get ticks everywhere
plt.minorticks_on()
plt.tick_params(top=True, right=True, direction='in', which='both')
plt.text(-7, 0, 'A', fontsize=ls,\
        horizontalalignment='center', verticalalignment='center')
plt.text(-16, 4, r'$\rm{A}^\prime$', fontsize=ls,\
        horizontalalignment='center', verticalalignment='center')
plt.text(10, -4, 'B', fontsize=ls,\
        horizontalalignment='center', verticalalignment='center')
plt.gca().get_xaxis().set_ticks(xtickvals)
plt.gca().get_yaxis().set_ticks(ytickvals)
#plt.xlabel('longitude (degrees)', fontsize=fs)
#plt.ylabel('latitude (degrees)', fontsize=fs)

t2 = a2.time[0]
vr_slice = a2.vals[:, :, idepth3, a2.lut[1], 0]/100

lon0 = 2*delta_phi_deflection_deg
iph1, iph2 = np.argmin(np.abs(lons - (lon0 - delta_lon/2))),\
    np.argmin(np.abs(lons - (lon0 + delta_lon/2)))
iph1, iph2 = np.argmin(np.abs(lons - (lon0 - delta_lon/2))),\
    np.argmin(np.abs(lons - (lon0 + delta_lon/2)))
lonvals = lons[iph1:iph2] - lon0
llons, llats = np.meshgrid(lonvals, latvals, indexing='ij')

patch = vr_slice[iph1:iph2, ith1:ith2]
plt.sca(ax2)
plt.pcolormesh(llons, llats, patch, cmap=plt.cm.RdYlBu_r, vmin=my_min, vmax=my_max, norm=MidpointNormalize(0))
plt.plot(lonvals, np.zeros_like(lonvals), 'k', linewidth=lw)
plt.plot(np.zeros_like(latvals), latvals, 'k', linewidth=lw)
# Get ticks everywhere
plt.minorticks_on()
plt.tick_params(top=True, right=True, direction='in', which='both')
plt.text(-6, 0, 'A', fontsize=ls,\
        horizontalalignment='center', verticalalignment='center')
plt.text(10, -4, 'B', fontsize=ls,\
        horizontalalignment='center', verticalalignment='center')
plt.gca().get_xaxis().set_ticks(xtickvals)
plt.gca().get_yaxis().set_ticks(ytickvals)
#plt.xlabel('longitude (degrees)', fontsize=fs)

t3 = a3.time[0]
vr_slice = a3.vals[:, :, idepth3, a3.lut[1], 0]/100

lon0 = 4*delta_phi_deflection_deg
iph1, iph2 = np.argmin(np.abs(lons - (lon0 - delta_lon/2))),\
    np.argmin(np.abs(lons - (lon0 + delta_lon/2)))
iph1, iph2 = np.argmin(np.abs(lons - (lon0 - delta_lon/2))),\
    np.argmin(np.abs(lons - (lon0 + delta_lon/2)))
lonvals = lons[iph1:iph2] - lon0
llons, llats = np.meshgrid(lonvals, latvals, indexing='ij')

patch = vr_slice[iph1:iph2, ith1:ith2]
plt.sca(ax3)
plt.pcolormesh(llons, llats, patch, cmap=plt.cm.RdYlBu_r, vmin=my_min, vmax=my_max, norm=MidpointNormalize(0))
plt.plot(lonvals, np.zeros_like(lonvals), 'k', linewidth=lw)
plt.plot(np.zeros_like(latvals), latvals, 'k', linewidth=lw)
# Get ticks everywhere
plt.minorticks_on()
plt.tick_params(top=True, right=True, direction='in', which='both')
plt.text(-6, 0, 'A', fontsize=ls,\
        horizontalalignment='center', verticalalignment='center')
plt.text(9, -4, 'B', fontsize=ls,\
        horizontalalignment='center', verticalalignment='center')
plt.gca().get_xaxis().set_ticks(xtickvals)
plt.gca().get_yaxis().set_ticks(ytickvals)
#plt.xlabel('longitude (degrees)', fontsize=fs)

t4 = a4.time[0]
vr_slice = a4.vals[:, :, idepth3, a4.lut[1], 0]/100

lon0 = 6*delta_phi_deflection_deg
iph1, iph2 = np.argmin(np.abs(lons - (lon0 - delta_lon/2))),\
    np.argmin(np.abs(lons - (lon0 + delta_lon/2)))
iph1, iph2 = np.argmin(np.abs(lons - (lon0 - delta_lon/2))),\
    np.argmin(np.abs(lons - (lon0 + delta_lon/2)))
lonvals = lons[iph1:iph2] - lon0
llons, llats = np.meshgrid(lonvals, latvals, indexing='ij')

patch = vr_slice[iph1:iph2, ith1:ith2]
plt.sca(ax4)
plt.pcolormesh(llons, llats, patch, cmap=plt.cm.RdYlBu_r, vmin=my_min, vmax=my_max, norm=MidpointNormalize(0))
plt.plot(lonvals, np.zeros_like(lonvals), 'k', linewidth=lw)
plt.plot(np.zeros_like(latvals), latvals, 'k', linewidth=lw)
# Get ticks everywhere
plt.minorticks_on()
plt.tick_params(top=True, right=True, direction='in', which='both')
plt.text(-8, 0, 'A', fontsize=ls,\
        horizontalalignment='center', verticalalignment='center')
plt.text(9, -4, 'B', fontsize=ls,\
        horizontalalignment='center', verticalalignment='center')
plt.text(-10, 16, 'C', fontsize=ls,\
        horizontalalignment='center', verticalalignment='center')
plt.text(-1, 0, 'D', fontsize=ls,\
        horizontalalignment='center', verticalalignment='center')
plt.gca().get_xaxis().set_ticks(xtickvals)
plt.gca().get_yaxis().set_ticks(ytickvals)
#plt.xlabel('longitude (degrees)', fontsize=fs)

cbar=plt.colorbar(cax=cbaxes, orientation='horizontal')
cbar.set_ticks([my_min, 0, my_max])
#cbar.ax.set_title('m/s', pad=10, horizontalalignment='left')

ax1.xaxis.set_ticklabels([])
ax2.xaxis.set_ticklabels([])
ax2.yaxis.set_ticklabels([])
ax4.yaxis.set_ticklabels([])

#ax8.xaxis.set_ticklabels(['-20', '10', '0', '10', '20'])

global_middle_x = margin_left + 0.5*(2*subplot_width + margin_middle_x)
global_middle_y = margin_bottom + 0.5*(2*subplot_height + margin_middle_y)
fig.text(0.4*margin_left, global_middle_y, 'latitude [deg]', rotation=90,\
         horizontalalignment='right', verticalalignment='center', fontsize=fs)
fig.text(global_middle_x, 0.5*margin_bottom, 'longitude [deg]', \
         horizontalalignment='center', verticalalignment='top', fontsize=fs)
# label colorbar

# Label the subplots
ax1.text(-18, 22, '(a)')
ax2.text(-18, 22, '(b)')
ax3.text(-18, 22, '(c)')
ax4.text(-18, 22, '(d)')

fig.text(cbar_left + cbar_width*1.1, cbar_bottom + cbar_height/2, 'm/s', verticalalignment='center')
savefile = '/nobackup/lmatilsk/rayleigh/paper_plots/n3_3_patch_evolve_4panels.png'
plt.savefig(savefile, dpi=300)
#plt.close()
```
