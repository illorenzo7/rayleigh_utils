```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

# Example data
t = np.arange(0.0, 1.0 + 0.01, 0.01)
s = np.cos(4 * np.pi * t) + 2

plt.plot(t, s)

plt.xlabel(r'\textbf{time} (s)')
plt.ylabel(r'\textit{voltage} (mV)',fontsize=16)
plt.title(r"\TeX\ is Number "
          r"$\displaystyle\sum_{n=1}^\infty\frac{-e^{i\pi}}{2^n}$!",
          fontsize=16, color='gray')
# Make room for the ridiculously large title.
plt.subplots_adjust(top=0.8)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
from diagnostic_reading import ReferenceState
from common import get_widest_range_file, strip_dirname

dir3 = '/castor/loma3853/rayleigh/n3_3'
datadir3 = dir3 + '/data/'

dir4 = '/sargas/loma3853/n4_3'
datadir4 = dir4 + '/data/'

dir5 = '/mizar/loma3853/n5_3'
datadir5 = dir5 + '/data/'

ref = ReferenceState(dir3 + '/reference', '')
rho = ref.density

# Get grid info (N3 and N4)
rr, tt, cost, sint, rr_depth, ri, ro, d = np.load(datadir3 + 'grid_info.npy')
nr, nt = len(rr), len(tt)

# Get reference state info
rho_2d = rho.reshape((1, nr))
r_2d = rr.reshape((1,nr))
sint_2d = sint.reshape((nt, 1))
rho_r_sint = rho_2d*r_2d*sint_2d

# Read in Reynolds stresses to compute amom_flux
rs_file3 = get_widest_range_file(datadir3, 'rs')
print ('Reading in Reynolds stress from ' + datadir3 + rs_file3 + ' ...')
(vr2_p, vt2_p, vp2_p, vrvp_p3, vrvt_p, vtvp_p,\
 vr2_m, vt2_m, vp2_m, vrvp_m3, vrvt_m, vtvp_m, fplus3, fminus3) =\
     np.load(datadir3 + rs_file3)

# Also need the average velocities to subtract off the component of the RS
# due to the meridional circulation
vavg_file3 = get_widest_range_file(datadir3, 'vavg')
vr_av3, vt_av3, vp_av3 = np.load(datadir3 + vavg_file3)

# Compute the radial and latitudinal angular momentum fluxes in the 
# meridional plane
amom_flux_r_p3 = rho_r_sint*(vrvp_p3 - fplus3*vr_av3*vp_av3)
amom_flux_r_m3 = rho_r_sint*(vrvp_m3 - fminus3*vr_av3*vp_av3)
amom_flux_r_t3 = amom_flux_r_p3 + amom_flux_r_m3
```

```python
# Case N4
# Read in Reynolds stresses to compute amom_flux
rs_file4 = get_widest_range_file(datadir4, 'rs')
print ('Reading in Reynolds stress from ' + datadir4 + rs_file4 + ' ...')
(vr2_p, vt2_p, vp2_p, vrvp_p4, vrvt_p, vtvp_p,\
 vr2_m, vt2_m, vp2_m, vrvp_m4, vrvt_m, vtvp_m, fplus4, fminus4) =\
     np.load(datadir4 + rs_file4)

# Also need the average velocities to subtract off the component of the RS
# due to the meridional circulation
vavg_file4 = get_widest_range_file(datadir4, 'vavg')
vr_av4, vt_av4, vp_av4 = np.load(datadir4 + vavg_file4)

# Compute the radial and latitudinal angular momentum fluxes in the 
# meridional plane
amom_flux_r_p4 = rho_r_sint*(vrvp_p4 - fplus3*vr_av4*vp_av4)
amom_flux_r_m4 = rho_r_sint*(vrvp_m4 - fminus3*vr_av4*vp_av4)
amom_flux_r_t4 = amom_flux_r_p4 + amom_flux_r_m4
```

```python
# case N5
ref5 = ReferenceState(dir5 + '/reference', '')
rho5 = ref5.density

# Get grid info (N5
rr5, tt5, cost5, sint5, rr_depth5, ri, ro, d = np.load(datadir5 + 'grid_info.npy')
nr5, nt5 = len(rr5), len(tt5)

# Get reference state info
rho_2d5 = rho5.reshape((1, nr5))
r_2d5 = rr5.reshape((1,nr5))
sint_2d5 = sint5.reshape((nt5, 1))
rho_r_sint5 = rho_2d5*r_2d5*sint_2d5

# Read in Reynolds stresses to compute amom_flux
rs_file5 = get_widest_range_file(datadir5, 'rs')
print ('Reading in Reynolds stress from ' + datadir5 + rs_file5 + ' ...')
(vr2_p, vt2_p, vp2_p, vrvp_p5, vrvt_p, vtvp_p,\
 vr2_m, vt2_m, vp2_m, vrvp_m5, vrvt_m, vtvp_m, fplus5, fminus5) =\
     np.load(datadir5 + rs_file5)

# Also need the average velocities to subtract off the component of the RS
# due to the meridional circulation
vavg_file5 = get_widest_range_file(datadir5, 'vavg')
vr_av5, vt_av5, vp_av5 = np.load(datadir5 + vavg_file5)

# Compute the radial and latitudinal angular momentum fluxes in the 
# meridional plane
amom_flux_r_p5 = rho_r_sint5*(vrvp_p5 - fplus5*vr_av5*vp_av5)
amom_flux_r_m5 = rho_r_sint5*(vrvp_m5 - fminus5*vr_av5*vp_av5)
amom_flux_r_t5 = amom_flux_r_p5 + amom_flux_r_m5
```

```python
# Average over low latitudes (+/- 15 degrees)

# case N3
# Theta grid in latitude degrees
tt_lat = tt*180/np.pi - 90
it1, it2 = np.argmin(np.abs(tt_lat - 15)), np.argmin(np.abs(tt_lat + 15))
flux_p3 = np.mean(amom_flux_r_p3[it1:it2+1, :], axis=0)
flux_m3 = np.mean(amom_flux_r_m3[it1:it2+1, :], axis=0)
flux_t3 = flux_p3 + flux_m3

# case N5
tt_lat5 = tt5*180/np.pi - 90
it1, it2 = np.argmin(np.abs(tt_lat5 - 15)), np.argmin(np.abs(tt_lat5 + 15))
flux_p5 = np.mean(amom_flux_r_p5[it1:it2+1, :], axis=0)
flux_m5 = np.mean(amom_flux_r_m5[it1:it2+1, :], axis=0)
flux_t5 = flux_p5 + flux_m5

savename = '/altair/loma3853/rayleigh/notebooks/amom_flux_lowlat.pdf'

# create plot
fig, axs = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(7.5, 3))
    
xmin, xmax = ri/ro, 1
xvals = np.linspace(xmin, xmax, 100)

plt.sca(axs[0])
plt.plot(rr/ro, flux_p3, 'b--', label='upflow')
plt.plot(rr/ro, flux_m3, 'b:', label='downflow')
plt.plot(rr/ro, flux_t3, 'b', label='total')
plt.xlim((xmin, xmax))
# Get ticks everywhere
plt.minorticks_on()
plt.tick_params(top=True, right=True, direction='in', which='both')
# Mark the zero line
plt.plot(xvals, np.zeros_like(xvals), 'k')
plt.xlabel(r'$r/r_o$', fontsize=14)
plt.ylabel(r'$\mathcal{F}_r\ (g/s^2)$', fontsize=14)

plt.legend()

plt.sca(axs[1])
plt.plot(rr5/ro, flux_p5, 'b--', label='upflow')
plt.plot(rr5/ro, flux_m5, 'b:', label='downflow')
plt.plot(rr5/ro, flux_t5, 'b', label='total')
plt.xlim((xmin, xmax))
# Get ticks everywhere
plt.minorticks_on()
plt.tick_params(top=True, right=True, direction='in', which='both')
# Mark the zero line
plt.plot(xvals, np.zeros_like(xvals), 'k')
plt.xlabel(r'$r/r_o$', fontsize=14)
plt.ylim((-1.5e16, 2e16))
plt.tight_layout()
plt.savefig(savename, dpi=300)
#plt.close()
```
