```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sys, os
sys.path.append('./rayleigh/plot')
from rayleigh_diagnostics import Shell_Slices
from matplotlib import ticker
from matplotlib import colors
from mpl_toolkits.basemap import Basemap, addcyclic
from common import get_widest_range_file, strip_dirname
from sslice_util import plot_ortho, units, labels, var_indices, cbar_axes_from_area
from binormalized_cbar import MidpointNormalize

# Get the run directory on which to perform the analysis
dir5 = '/altair/loma3853/rayleigh/n5_sslice_19100050'
radatadir = dir5 + '/vr/'
iter0 = 19100050

# Directory to save plots
datadir = dir5 + '/data/'
plotdir = dir5 + '/plots/'

if not os.path.isdir(plotdir):
    os.makedirs(plotdir)
```

```python
# Get grid info
rr,tt,cost,sint,rr_depth,ri,ro,d = np.load(datadir + 'grid_info.npy')
nt, nr = len(tt), len(rr)
nphi = 2*nt
cost_nd = cost.reshape((1, nt))
sint_nd = sint.reshape((1, nt))
phi = np.arange(0, 2*np.pi, 2*np.pi/nphi)
lats = 90 - tt*180/np.pi 
lons = phi*180/np.pi - 180

# Set dimensions/location of patch to plot 
delta_lat = 40
delta_lon = 40
lon0 = 0
lat0 = 0

ith1, ith2 = np.argmin(np.abs(lats - (lat0 - delta_lat/2))),\
    np.argmin(np.abs(lats - (lat0 + delta_lat/2)))

iph1, iph2 = np.argmin(np.abs(lons - (lon0 - delta_lon/2))),\
    np.argmin(np.abs(lons - (lon0 + delta_lon/2)))
    
xtickvals = [lon0 - delta_lon/2, lon0 - delta_lon/4, lon0, lon0 + delta_lon/4, lon0 + delta_lon/2]
ytickvals = [lat0 - delta_lat/2, lon0 - delta_lat/4, lat0, lat0 + delta_lat/4, lat0 + delta_lat/2]

lonvals = lons[iph1:iph2]
latvals = lats[ith1:ith2]
llons, llats = np.meshgrid(lonvals, latvals, indexing='ij')
```

```python
# Read shell slice for panels
print ('Reading ' + radatadir + str(iter0).zfill(8) + ' ...')
a = Shell_Slices(radatadir + str(iter0).zfill(8), '')
vr = a.vals[:, :, :, 0, 0]/100
```

```python
# Read in patches at different depths
min_depth = 0.05
max_depth = 0.5

desired_depths = np.linspace(min_depth, max_depth, 8)

alldepths = (ro - a.radius)/d

depth_indices = []
for desired_depth in desired_depths:
    depth_indices.append(np.argmin(np.abs(alldepths - desired_depth)))
    
rrvals = a.radius[depth_indices]
rrnvals = rrvals/ro

patches = vr[iph1:iph2, ith1:ith2, depth_indices]
```

```python
def rms(array):
    if np.size(array) == 0:
        return 0
    else:
        return np.sqrt(np.mean(array**2))
```
